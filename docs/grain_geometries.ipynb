There are many possibilities of grain geometries in SolidPy. Besides the simple Bates grain, SolidPy can create regression maps for Star ports. There is a predefined Star geometry class that receives the arguments: `outer_radius`, `star_minimum`, `star_maximum` and `star_points`.

```python
def radius(t):
    m=3
    n=5
    k=1
    r = np.cos((2*np.arcsin(k)+np.pi*m)/(2*n)) / np.cos((2*np.arcsin(k*np.cos(n*t))+np.pi*m)/(2*n))
    return r

Custom = CustomGeometry(
    outer_radius=2,
    inner_radius=radius,
    height=1,
    input_method="polar",
)
```
